```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('board2.csv') # load the dataset from the 'board2.csv'
df 
```

#### a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
Instru = df[(df['Track'] == 'Instrumentation') & (df['Hometown'] == 'Luzon')]  #filters df for students in Instrumentation from Luzon with Electronics score above 70 to select specific columns
Instru = Instru[['Name', 'GEAS', 'Electronics']]
Instru = Instru[Instru['Electronics'] > 70]
Instru
```

```python
#load the dataset from 'board2.csv' into a DataFrame
#compute the average score for each student by averaging the 'Math', 'Electronics', 'GEAS', and 'Communication' columns
#add the calculated average score as a new column 'Average' in the DataFrame


df = pd.read_csv('board2.csv')
df['Average'] = (df['Math']+df['Electronics']+df['GEAS']+df['Communication'])/4
df
```

#### b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
Instru = df.loc[((df['Hometown']== 'Mindanao')&(df['Gender']== 'Female')&(df['Average']>=55)), ['Name', 'Track', 'Electronics']]
Instru
```

#### Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
#sets the figure size for the plot
#creates a bar plot to visualize the average scores for each track
#labels the x-axis as 'Track' and the y-axis as 'Average'
#adds a title to the plot


plt.figure (figsize = (7,9))
plt.bar(df ['Track'], df ['Average'])
plt.ylabel ("Average", fontsize = 20)
plt.xlabel ("Track", fontsize = 20)
plt.title ("Table 1. The relationship between the track and average graph", fontsize = 30)
plt.show()
```

```python
#sets the figure size for the plot
#creates a bar plot to visualize the average scores for each hometown
#labels the x-axis as 'Hometown' and the y-axis as 'Average' 
#adds a title to the plot



plt.figure (figsize = (7,9))
plt.bar(df ['Hometown'], df ['Average'])
plt.ylabel ("Average", fontsize = 20)
plt.xlabel ("Hometown", fontsize = 20)
plt.title ("Table 2. The relationship between the hometown and average graph", fontsize = 30)
plt.show()
```

```python
#set the figure size for the plot
#creates a bar plot to visualize the average scores for each gender
#labels the x-axis as 'Gender' and the y-axis as 'Average' 
#adds a title to the plot


plt.figure (figsize = (7,9))
plt.bar(df ['Gender'], df ['Average'])
plt.ylabel ("Average", fontsize = 20)
plt.xlabel ("Gender", fontsize = 20)
plt.title ("Table 2. The relationship between the Gender and average graph", fontsize = 30)
plt.show()
```
